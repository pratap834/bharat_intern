```python
columns = ['sepal_l', 'sepal_w', 'petal_l', 'petal_w', 'class']
df.columns = columns
```

```python
sns.pairplot(df, hue=df.columns.all())
plt.show()
```

```python
from sklearn.model_selection import train
```

```python
data = df.values  #return numpy array
#var = data[y-axis, x-axis]
x = data[:, 0:4]
y = data[:, 4]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=101)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(x_train, y_train)
```

```python
svc_prediction = svc_model.predict(x_test)
```

```python
svc_acc = accuracy_score(y_test, svc_prediction)
```

```python
print(f'Accuracy: {svc_acc}')
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
lr_acc = accuracy_score(y_test, prediction)
```

```python
print(f'Accuracy: {lr_acc}')
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, prediction)
print(report)
```
